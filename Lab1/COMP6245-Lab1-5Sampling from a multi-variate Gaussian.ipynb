```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
# numpy.linalg.cholesky(a) is cholesky decomposition. More information click link below
# https://numpy.org/doc/stable/reference/generated/numpy.linalg.cholesky.html
C = np.array([[0.5,6],[1,4]])
A = np.linalg.cholesky(C)
print(A)
print(A @ A.T)
```

```python
X = np.random.randn(1000,2)
Y = X @ A.T
print(X.shape)
print(Y.shape)
```

```python
fig, ax = plt.subplots(figsize=(5,5))
ax.scatter(X[:,0], X[:,1], c="c", s=4)
ax.scatter(Y[:,0], Y[:,1], c="m", s=4)
ax.set_xlim(-6, 6)
ax.set_ylim(-6, 6)
```
