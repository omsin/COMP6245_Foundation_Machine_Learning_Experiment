```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python

x = np.random.rand(1000,1)
# figsize is overall size of figure
fig, ax = plt.subplots(nrows=1, ncols=2,figsize=(10,5))
n1bins, n2bins = 10, 100
ax[0].hist(x, bins=n1bins)
ax[0].set_ylim(0,1200)
ax[0].set_xlabel("Bins", fontsize=16)
ax[0].set_ylabel("Count", fontsize=16)
ax[0].tick_params(axis='both', which='major', labelsize=14)
ax[0].set_title("Histogram: 10000 sample, bins=%4d"%(n1bins), fontsize=16)
ax[1].hist(x, bins=n2bins)
ax[1].set_ylim(0,120)
ax[1].set_xlabel("Bins", fontsize=16)
ax[1].set_ylabel("Count", fontsize=16)
ax[1].tick_params(axis='both', which='major', labelsize=14)
ax[1].set_title("Histogram: 10000 sample, bins=%4d"%(n2bins), fontsize=16)
plt.savefig("histograms_uniform.png")
plt.tight_layout()
```

```python
# numpy.zeros(shape, dtype=float, order='C') is return a new array of given shape and type, filled with zeros
N = 1000
x1 = np.zeros(N)
for n in range(N):
    x1[n] = np.sum(np.random.rand(12,1)) - np.sum(np.random.rand(12,1))
    print(x1[n])
fig, ax = plt.subplots(figsize=(5,5))
ax.hist(x1, 20)
```
